# Time Complexity Comparison

---
## Time complexity of bico(n, i)

The time complexity of int bico(int n, int i) is O(n^2). This is because the function iterativePascal(n) creates a 2D array of size n x n, and the nested loops in that function iterate n*(n+1)/2 times to fill in the triangle. The subsequent loop in bico(n, i) iterates n+1 times to delete the 2D array. Therefore, the time complexity of iterativePascal(n) is O(n^2), and the time complexity of bico(n, i) is also O(n^2) because it calls iterativePascal(n).

---
## Time complexity of recursiveBico(n, i)

The time complexity of int recursiveBico(int degree, int index) is exponential, specifically O(2^n), where n is the degree of the polynomial. This is because each recursive call branches into two more recursive calls, leading to a binary tree with 2^n nodes. Therefore, the total number of recursive calls made by the function is proportional to 2^n. This makes the recursive implementation of Pascal's triangle less efficient fro greater data sizes than the iterative implementation, which has a time complexity of O(n^2).


---
## Comparison

The time complexity of bico() and recursiveBico() functions differ significantly. The iterative approach of bico() function uses a precomputed Pascal's triangle to compute the binomial coefficient, taking constant time O(1) to return the coefficient. Therefore, the time complexity of bico() function is O(n^2), where n is the degree of the polynomial. On the other hand, recursiveBico() function uses a recursive implementation of Pascal's triangle to compute the binomial coefficient. It recursively calls itself, computing the coefficient by summing the results of two recursive calls. Therefore, the time complexity of recursiveBico() function is O(2^n), where n is the degree of the polynomial. This means that the recursiveBico() function is an exponential time algorithm, which makes it impractical for large values of n. Although the recursive approach has a simpler implementation, it is not suitable for large polynomial degrees. In contrast, the iterative approach used in bico() function provides a fast and efficient algorithm for computing binomial coefficients of moderate-sized polynomials.

---
## Conclusion

In conclusion, the time complexity of computing binomial coefficients using Pascal's triangle depends on the method used. While the iterative approach has a time complexity of O(n^2), the recursive approach has an exponential time complexity of O(2^n). Therefore, for large values of n, the recursive approach can be significantly slower than the iterative approach.
